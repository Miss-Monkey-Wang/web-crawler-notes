# 东方财富网龙虎榜数据分析

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
pd.set_option('display.max_columns',100)
```

```python
url = 'http://data.eastmoney.com/DataCenter_V3/stock2016/TradeDetail/pagesize=200,page=1,sortRule=-1,sortType=,startDate=2019-04-09,endDate=2019-04-09,gpfw=0,js=var%20data_tab_1.html?rt=25898785'
```

```python
r = requests.get(url)
pat = '"data":(\[.*?\])'
data = re.compile(pat,re.S).findall(r.text)
data=pd.DataFrame(eval(data[0]))
# web_content.encoding='utf-8'
# bs = BeautifulSoup(web_content.text)
```

```python
col_map = {
 'Bmoney': '龙虎榜买入额(万)',
 'Chgradio': '涨跌幅',
 'ClosePrice': '收盘价',
 'Dchratio': '换手率',
 'JmMoney': '龙虎榜净买额(万)',
 'JmRate': '净买额占总成交比',
 'Ltsz': '流通市值(亿)',
 'Rchange10dc': '上榜后10日',
 'Rchange1dc': '上榜后1日',
 'Rchange2dc': '上榜后2日',
 'Rchange5dc': '上榜后5日',
 'SCode': '代码',
 'Smoney': '龙虎榜卖出额(万)',
 'Tdate': '上榜日',
 'Turnover': '市场总成交额(万)',
 'ZeMoney': '龙虎榜成交额(万)',
 'ZeRate': '成交额占总成交比',
 'SName': '名称',
 'JD' : '解读',
 'Ctypedes': '上榜原因'}
```

```python
sorted_cols = ['代码','名称','解读','收盘价','涨跌幅','龙虎榜净买额(万)','龙虎榜买入额(万)',
 '龙虎榜卖出额(万)','龙虎榜成交额(万)','市场总成交额(万)','净买额占总成交比','成交额占总成交比',
 '换手率','流通市值(亿)','上榜原因', 
#  '上榜后10日',
#  '上榜后1日',
#  '上榜后2日',
#  '上榜后5日', 
 '上榜日']
```

```python
amt_w_cols = ['龙虎榜净买额(万)','龙虎榜买入额(万)',
 '龙虎榜卖出额(万)','龙虎榜成交额(万)','市场总成交额(万)']

amt_y_cols = ['流通市值(亿)']
```

```python
data.columns = [ col_map.get(c,c) for c in data.columns]
```

```python
data=data.loc[:,sorted_cols]
```

```python
for c in data.columns:
    try:
        data[c]=data[c].astype('float')
    except:
        print(c)
```

```python
for c in amt_w_cols:
    data[c]= round(data[c]/10000,2)
for c in amt_y_cols:
    data[c]=round(data[c]/100000000,2)
```
